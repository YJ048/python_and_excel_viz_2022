```python
#page 69 - 70
#labeling the plots and function

step = 1
x = np.arange(0, 5 + step, step) # an array 0,1,2,3,4,5
y = np.exp(x) # an array of exp(0), exp(1), ..., exp(5)

plt.plot(x,y,
        linewidth=3,
        linestyle=':',
        color='r',
        label='Line 1')
plt.plot(x,y+30,
        linewidth=2,
        linestyle='-.',
        color='b',
        label='Line 2')
plt.plot(x,y+60,
        marker='o',
        linestyle='--',
        color='g',
        label='Line 3')

plt.legend(fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
## page 64 - 68
## figure graph

step = 1
x = np.arange(0, 10+step, step)
y = np.exp(x/2) 

plt.figure(figsize=(10, 7)) # size of the figure also
plt.title('My Title', fontsize=15) #can add title
plt.bar(x,y,
       width=0.6,
       color='b',
       alpha=0.7)
plt.xlabel('x-axis',fontsize=12) #change font size
plt.ylabel('y-axis',fontsize=12) #change font size
plt.show()
```

```python
array_2d = np.array([[18,26,17],
                     [25,15.5,12],
                     [24,27,20],
                     [10,5.5,17],
                     [27,26,15],
                     [22,21,21],])

print("Before\n", array_2d)
print("Shape of Array is ",array_2d.shape)
for r in range(array_2d.shape[0]): # for r in range(6) -> 0,1,2,3,4,5
    for c in range(array_2d.shape[1]): # for c in range(3) -> 0,1,2
        array_2d[r,c] = array_2d[r,c] + 1 # add 1 to the element at (row r, column c)

print("After \n",array_2d)
```
